# Pass by reference vs Value


```python
# function definition is here
def changeme(mylist):
    "this changes a passed list into this function"
    mylist.append([1,2,3,4]);
    print("values inside the function: ", mylist)
    return

#now you can call printme function
mylist=[10,20,30];
changeme(mylist);
print("values outside the function: ", mylist)
```

```python
# argumen diteruskan oleh reference dan reference kemudian dioverwrite didalam fungsi yang dipanggil
# function definition is here
def changeme(mylist):
    "this changes a passed list into this function"
    mylist=([1,2,3,4]);
    print("values inside the function: ", mylist)
    return

#now you can call printme function
mylist=[10,20,30];
changeme(mylist);
print("values outside the function: ", mylist)
```

##### buat function yang berisi 3 parameter string yang akan dijoin menggunakan spasi dan diconvert menjadi reverse

* contoh :
* input :
* str_joiner('Saya','Mau','Makan')
* output:
* nakam uam ayas

```python
def str_joiner(a,b,c):
    a=a.lower()
    b=b.lower()
    c=c.lower()
    return (a + ' ' + b + ' '+ c) [::-1]
str_joiner('Saya','Mau','Makan')
```

# Variabel length arguments

```python
def printinfo(x, *vartuple):
    print('Output is :')
    print(x)
    print(vartuple)
    
printinfo(8,9,20,30,'Nama',90,100,11)
```

```python
def printinfo(x, *vartuple):
    print('Output is :')
    print(x)
    for i in vartuple:
        print(i)
    
printinfo(8,9,20,30,'Nama',90,100,11)
```
